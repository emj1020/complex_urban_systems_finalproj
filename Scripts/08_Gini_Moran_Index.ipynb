```python
def get_gini(tt_df,geo_colnm):
    if geo_colnm == None:
        N = len(tt_df)
        diff_weighted_mat = np.zeros((N,N))
        diff_mean_mat = np.zeros((N,N))
        diff_min_mat = np.zeros((N,N))
        
        for i in range(N):
            for j in range(N):
                diff_weighted_mat[i,j] = abs(tt['w_tt'][i]-tt['w_tt'][j])
                diff_mean_mat[i,j] = abs(tt['mean_tt'][i]-tt['mean_tt'][j])
                diff_min_mat[i,j] = abs(tt['min_tt'][i]-tt['min_tt'][j])
        
        double_sum_weighted = np.sum(np.sum(diff_weighted_mat,axis = 1))
        double_sum_mean = np.sum(np.sum(diff_mean_mat,axis = 1))
        double_sum_min = np.sum(np.sum(diff_min_mat,axis = 1))

        w_tt_mean = tt['w_tt'].mean()
        mean_tt_mean = tt['mean_tt'].mean()
        min_tt_mean = tt['min_tt'].mean()
        
        Gini_w = double_sum_weighted/(2*N**2*w_tt_mean)
        Gini_mean = double_sum_mean/(2*N**2*mean_tt_mean)
        Gini_min = double_sum_min/(2*N**2*min_tt_mean)
    else:
        print('Successfully made it to else statement')
        Gini_w = []
        Gini_mean = [] 
        Gini_min = []
        for geo in tt_df[geo_colnm].unique():
            #iterate through unique geographies and calculate gini indexes
            subset_df = tt_df[tt_df[geo_colnm] == geo].reset_index(drop = True)
            N = len(subset_df)
            diff_weighted_mat = np.zeros((N,N))
            diff_mean_mat = np.zeros((N,N))
            diff_min_mat = np.zeros((N,N))
            
            for i in range(N):
                for j in range(N):
                    diff_weighted_mat[i,j] = abs(subset_df['w_tt'][i]-subset_df['w_tt'][j])
                    diff_mean_mat[i,j] = abs(subset_df['mean_tt'][i]-subset_df['mean_tt'][j])
                    diff_min_mat[i,j] = abs(subset_df['min_tt'][i]-subset_df['min_tt'][j])

            double_sum_weighted = np.sum(np.sum(diff_weighted_mat,axis = 1))
            double_sum_mean = np.sum(np.sum(diff_mean_mat,axis = 1))
            double_sum_min = np.sum(np.sum(diff_min_mat,axis = 1))

            w_tt_mean = subset_df['w_tt'].mean()
            mean_tt_mean = subset_df['mean_tt'].mean()
            min_tt_mean = subset_df['min_tt'].mean()

            Gini_w.append(double_sum_weighted/(2*N**2*w_tt_mean))
            Gini_mean.append(double_sum_mean/(2*N**2*mean_tt_mean))
            Gini_min.append(double_sum_min/(2*N**2*min_tt_mean))
    return Gini_w,Gini_mean,Gini_min
```

## Import geography boundaries from NYC Open Data Portal

* Borough Boundaries: `https://data.cityofnewyork.us/resource/7t3b-ywvw.json`

* PUMA Boundaries: `https://data.cityofnewyork.us/resource/ith2-q832.json`

* NTA Boundaries: `https://data.cityofnewyork.us/resource/q2z5-ai38.json`

```python
#import nta shapefile
urls = ['https://data.cityofnewyork.us/resource/7t3b-ywvw.json',
        'https://data.cityofnewyork.us/resource/ith2-q832.json',
        'https://data.cityofnewyork.us/resource/q2z5-ai38.json']
query = (urls[0]+'?'
        "$select=*"
        "&$limit=10000")
query = query.replace(" ", "%20")
response = urllib.request.urlopen(query)
data = json.loads(response.read())


for d in data:
    d['geometry'] = shape(d['the_geom'])

#store in a geodataframe
boro = gpd.GeoDataFrame(data, geometry = 'geometry', crs = 'EPSG: 4326')
boro = boro.drop(columns = ['the_geom'])
boro.head()
```

# Computing spatial auto-correlation or Moran's I

We are going to use pysal package in python to compute the spatial weights of our travel times across Census tracts in NYC. 

**Data:**

* 2010 Census Tracts: `../Data/2010_Census_Tracts/2010_Census_Tracts.shp`

**Sources:**

* [Local Spatial Autocorrelation in Python](https://geographicdata.science/book/notebooks/07_local_autocorrelation.html)

```python
from matplotlib import colors
import seaborn  # Graphics
import rioxarray  # Surface data manipulation
import xarray  # Surface data manipulation
from pysal.explore import esda  # Exploratory Spatial analytics
from pysal.lib import weights  # Spatial weights
import contextily  # Background tiles
```

```python
tracts_gdf = gpd.read_file('../Data/2010_Census_Tracts/2010_Census_Tracts.shp')
tracts_gdf.head()
```

```python
tracts_gdf['geoid'] = tracts_gdf['geoid'].astype(int) 
tt['geoid'] = tt['geoid'].astype(int)

tracts_tt = tracts_gdf.merge(tt[['geoid','Bus_Access','Subway_Access',
                                 'Bus_Count', 'Subway_Count',
                                'w_tt', 'mean_tt', 'min_tt']],on = 'geoid',how = 'left')
tracts_tt.head()
```

```python
tracts_tt.columns
```

```python
#create dataframes with only necessary columns
min_tt_gdf = tracts_tt[['geoid','geometry','min_tt']]
w_tt_gdf = tracts_tt[['geoid','geometry','w_tt']]
```

```python
len(min_tt_gdf[min_tt_gdf['min_tt'].isna()])
```

```python
len(w_tt_gdf[w_tt_gdf['w_tt'].isna()])
```

```python
#drop nan values
min_tt_gdf = min_tt_gdf.dropna()
w_tt_gdf = w_tt_gdf.dropna()

print(len(min_tt_gdf[min_tt_gdf['min_tt'].isna()]),len(w_tt_gdf[w_tt_gdf['w_tt'].isna()]))
```

```python
# Generate W from the GeoDataFrame
w_min_tt = weights.distance.KNN.from_dataframe(min_tt_gdf, k=8)
w_w_tt = weights.distance.KNN.from_dataframe(w_tt_gdf, k=8)
# Row-standardization
w_min_tt.transform = "R"
w_w_tt.transform = "R"
```

```python
#calculate spatial lag
min_tt_gdf["w_min_tt"] = weights.lag_spatial(w_min_tt, min_tt_gdf['min_tt'])
w_tt_gdf["w_w_tt"] = weights.lag_spatial(w_w_tt, w_tt_gdf['w_tt'])
```

```python
#standard deviations
min_tt_gdf["min_tt_std"] = min_tt_gdf['min_tt'] - min_tt_gdf['min_tt'].mean()
min_tt_gdf["w_min_tt_std"] = weights.lag_spatial(w_min_tt, min_tt_gdf['min_tt_std'])

w_tt_gdf["w_tt_std"] = w_tt_gdf['w_tt'] - w_tt_gdf['w_tt'].mean()
w_tt_gdf["w_w_tt_std"] = weights.lag_spatial(w_w_tt, w_tt_gdf['w_tt_std'])
```

```python
# Set up the figure and axis
fig, ax = plt.subplots(1,2, figsize=(10, 6))
# Plot values
seaborn.regplot(x="min_tt_std", y="w_min_tt_std", data=min_tt_gdf, ci=None,ax = ax[0])
seaborn.regplot(x="w_tt_std", y="w_w_tt_std", data=w_tt_gdf, ci=None,ax = ax[1])

# Add vertical and horizontal lines
ax[0].axvline(0, c="k", alpha=0.5)
ax[0].axhline(0, c="k", alpha=0.5)
# Add text labels for each quadrant
ax[0].text(25, 5, "HH", fontsize=25, c="r")
ax[0].text(17, -11, "HL", fontsize=25, c="r")
ax[0].text(-20, 8.0, "LH", fontsize=25, c="r")
ax[0].text(-25, -11.0, "LL", fontsize=25, c="r")
ax[0].set_title('Minimum Commute Time')
ax[1].axvline(0, c="k", alpha=0.5)
ax[1].axhline(0, c="k", alpha=0.5)
# Add text labels for each quadrant
ax[1].text(25, 5, "HH", fontsize=25, c="r")
ax[1].text(17, -11, "HL", fontsize=25, c="r")
ax[1].text(-20, 8.0, "LH", fontsize=25, c="r")
ax[1].text(-27, -11.0, "LL", fontsize=25, c="r")
ax[1].set_title('Wieghted Commute Time')
plt.savefig('../Outputs/min_wieghted_quardrants.png')
plt.show()
```

```python
morani_mintt = esda.moran.Moran_Local(min_tt_gdf['min_tt'], w_min_tt)
morani_wtt = esda.moran.Moran_Local(w_tt_gdf['w_tt'], w_w_tt)
```

```python
from splot import esda as esdaplot

# Set up figure and axes
f, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
# Make the axes accessible with single indexing
axs = axs.flatten()

# Subplot 1 #
# Choropleth of local statistics
# Grab first axis in the figure
ax = axs[0]
# Assign new column with local statistics on-the-fly
min_tt_gdf.assign(
    Is=morani_mintt.Is
    # Plot choropleth of local statistics
).plot(
    column="Is",
    cmap="plasma",
    scheme="quantiles",
    k=5,
    edgecolor="white",
    linewidth=0.1,
    alpha=0.75,
    legend=True,
    ax=ax,
)

# Subplot 2 #
# Quadrant categories
# Grab second axis of local statistics
ax = axs[1]
# Plot Quadrant colors (note to ensure all polygons are assigned a
# quadrant, we "trick" the function by setting significance level to
# 1 so all observations are treated as "significant" and thus assigned
# a quadrant color
esdaplot.lisa_cluster(morani_mintt, min_tt_gdf, p=1, ax=ax)

# Subplot 3 #
# Significance map
# Grab third axis of local statistics
ax = axs[2]
#
# Find out significant observations
labels = pd.Series(
    1 * (morani_mintt.p_sim < 0.05),  # Assign 1 if significant, 0 otherwise
    index=min_tt_gdf.index  # Use the index in the original data
    # Recode 1 to "Significant and 0 to "Non-significant"
).map({1: "Significant", 0: "Non-Significant"})
# Assign labels to `db` on the fly
min_tt_gdf.assign(
    cl=labels
    # Plot choropleth of (non-)significant areas
).plot(
    column="cl",
    categorical=True,
    k=2,
    cmap="Paired",
    linewidth=0.1,
    edgecolor="white",
    legend=True,
    ax=ax,
)


# Subplot 4 #
# Cluster map
# Grab second axis of local statistics
ax = axs[3]
# Plot Quadrant colors In this case, we use a 5% significance
# level to select polygons as part of statistically significant
# clusters
esdaplot.lisa_cluster(morani_mintt, min_tt_gdf, p=0.05, ax=ax)

# Figure styling #
# Set title to each subplot
for i, ax in enumerate(axs.flatten()):
    ax.set_axis_off()
    ax.set_title(
        [
            "Local Statistics",
            "Scatterplot Quadrant",
            "Statistical Significance",
            "Moran Cluster Map",
        ][i],
        y=0,
    )
# Tight layout to minimize in-between white space
f.suptitle('Minimum Commute Time Spatial Statistics',fontsize = 15)
f.tight_layout()
plt.savefig('../Outputs/min_spatial_stats.png',bbox_inches = 'tight')
# Display the figure
plt.show()
```
