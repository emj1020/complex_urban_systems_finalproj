```python
import pandas as pd
import numpy as np
```

```python
tt = pd.read_csv('../Data/commute_tt_results_v2.csv')
tt.head()
```

```python
#column for mean 
tt['mean_to_cbd'] = tt.iloc[:, -22:].mean(axis=1)
```

```python
#travel time to the closest cbd
tt['closest_cbd'] = tt.iloc[:, -22:].min(axis=1)
```

```python
#load cbd data 
cbd = pd.read_csv('../Data/NTA_CBD_final.csv')
cbd
```

```python
#multiply %job, mean, min, and original value(?)
for i in range(22):
    tt['w_tt_{}'.format(i+1)] = tt.loc[:, 'CBD{}'.format(i+1)]*cbd['%_cbd_jobs'][i]*tt['closest_cbd']*tt['mean_to_cbd']
tt
```
